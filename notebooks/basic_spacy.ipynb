```python
from spacy.matcher import PhraseMatcher

matcher = PhraseMatcher(nlp.vocab)
terms = ["apache beam", "PCollection", "PTransform", "Aggregation", "data processing pipeline"]
patterns = [nlp.make_doc(text) for text in terms]
matcher.add("TerminologyList", patterns)


doc = nlp("An apache beam pipeline is a graph of transformations that defines the data processing pipeline")

matches = matcher(doc)

for match_id, start, end in matches:
    span = doc[start:end]
    print(span.text)
```

```python
from spacy import displacy 

doc = nlp("An apache beam pipeline is a graph of transformations that defines the data processing pipeline")
displacy.serve(doc, style="dep")
```

```python
import pandas as pd 
def table_print(doc):
    data = []
    for token in doc:
        data.append([token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
                token.shape_, token.is_alpha, token.is_stop])
    return pd.DataFrame(data=data, columns=["TEXT", "LEMMA", "POS", "TAG", "DEP", "SHAPE", "ALPHA", "STOP"])
            

table_print(doc)
    
```

```python
doc = nlp("A window in apache beam subdivides a PCollection based on timestamps")
table_print(doc)
```

```python
doc = nlp("An aggregation in apache beam computes a value from multiple elements")
table_print(doc)
```
